```python
import pandas as pd
```

```python
ncbi_gene = pd.read_csv("Mammalia/Homo_sapiens.gene_info.tsv", sep="\t", index_col=1)
ncbi_gene.head()
```

```python
genes = pd.read_csv("../csv/Genes.nodes.csv", index_col=0)
genes.head() 
```

```python
new_genes = [] 
for gene_id, row in genes.iterrows():
	data = {
		"id": gene_id,
		**row
	}
	if gene_id in ncbi_gene.index:
		dbXrefs = ncbi_gene.loc[gene_id, 'dbXrefs']
		for  i in dbXrefs.split("|"):
			key, *val = i.split(":")
			val = ":".join(val)
			if key in ['HGNC', 'Ensembl']:
				data[key] = val
	new_genes.append(data)
new_genes = pd.DataFrame.from_dict(new_genes).set_index('id')
new_genes.head()
```

```python
new_genes = new_genes[[i for i in new_genes.columns if not i == 'uri']]
```

```python
new_genes.to_csv('../csv/Genes.nodes.csv')
```
